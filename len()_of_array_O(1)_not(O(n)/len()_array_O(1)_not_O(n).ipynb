# Explanation why an array is pre-determined its length at initialization.



- Both strings and arrays (or lists) in Python store their length as part of their internal data structure, which allows for the len() function to operate in constant time O(1).

**Arrays (or Lists) in Python**
- **Stored Length**: Just like strings, arrays (or lists) in Python also store their length as an **attribute**. This allows the length to be quickly accessed without having to traverse the entire array.
-**Memory Layout**: Both arrays (lists) and strings are implemented as objects in Python, and they have a structure that includes a field for the length of the object.
