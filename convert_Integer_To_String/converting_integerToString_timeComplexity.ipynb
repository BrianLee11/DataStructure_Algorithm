# Motivation
- In Python, str() converts to a string data structure.
- Instead of using str(), write an algorithm to resemble the output of str()
- Show time complexity of this algorithm.

# Example
- integer = 10
- str(integer) = ['1', '0']

`````python
class Solution(object):
  def convertIntegerToString(self, integer):
    """
      input: a non-negative integer
      output: a string with characters in the integer
      intuition: Use % and // to iterate each digit in the given integer. Then reverse its characters. Then join the characters.
      time complexity: O(digits) ~ O(log_{10}(integer))
    """

    # initialize a string to store characters from integer
    string = []

    # iterate each digit in a reverse order, and attach in array
    # example: 12345 -> [5, 4, 3, 2, 1]
    number = integer
    divisor = 10
    reversed_string = []

    while integer != 0:
      remainder = integer % 10    # Get the last digit
      string.append(chr(remainder + ord('0')))  # Convert digit to character
      integer = integer // 10      # Remove the last digit

    # reverse the resversed_string
    # example: [5, 4, 3, 2, 1] -> [1, 2, 3, 4, 5]

    for index in range(len(reversed_string)):
      string.append(reversed_string[index])

    # join characters within the array
    # example: [1,2,3,4,5] -> [12345]
    return ''.join(string)
`````

# note about remainder + ord('0')

- ord('0'): The ord function returns the Unicode code point (an integer) of the character '0', which is 48. This is used as a base to convert numeric digits to their corresponding character codes.

- remainder + ord('0'): By adding ord('0') to remainder, we effectively map the numeric digit to its corresponding character code. For example:

- If remainder is 5, then remainder + ord('0') is 5 + 48 which equals 53.
The Unicode code point 53 corresponds to the character '5'.


`````python
# example
solution = Solution()

integer = 1234512345

result = solution.convertIntegerToString(integer)
print(result)
`````

# Analysis: time complexity is $O_{10}(\text{integer})$

1. **Initialization:" $O(1)$
```
    # initialize a string to store characters from integer
    string = []

    # iterate each digit in a reverse order, and attach in array
    # example: 12345 -> [5, 4, 3, 2, 1]
    number = integer
    divisor = 10
    reversed_string = []
```

2. **Interation part: the while loop**

```
    while number != 0:
      remainder = number % 10    # example: 12345 % 10 = 5
      reversed_string.append(remainder)
      number = number // 10      # example: 12345 // 10 =  1234
```

- The while loop `number != 0` takes $O(d)$, where d = number of digits in the integer. Example: integer = 12345, d = 5. This takes $O(5)$.  
- `remainder = number % 10` takes $O(1)$ since this is a constant operation.
- `reversed_string.append(remainder)` also takes $O(1)$ for the same reason as above.
- `number = number // 10` also takes $O(1)$.

- To approximate $d$, we would take log of $integer$ with respect to 10, $$O(d) = O(log_{10} (\text{integer}))$$


3. **Reverse the elements in the array**
```
    for index in range(len(reversed_string)):
      string.append(reversed_string[index])    
```
- This would be also approximately $O(d) = O(log_{10} (integer))$


4. **Adding Negative Sign and Joining the List:**
```
''.join(string)
````
- $O(d)$ as well.

**Conclusion:** In overall, since the dominant time complexity is $O(d)$, this algorithm takes $$O(d) = O(log_{10} (\text{integer}))$$
