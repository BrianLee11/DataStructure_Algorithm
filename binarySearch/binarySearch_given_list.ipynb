# **Binary search algorithm**
Creation date: Saturday, July 6, 2024

**Motivation:** rather than visiting every single element of the given list (of numbers), which would be $O(n)$, binary search allows $O(log_{2}(n))$ to search the target number.

- **Input**: 1. list of numbers, 2. a target number within this list of numbers.  

- **Output**: target number.

- **Assumption**: The list of numbers has equal distance for each number (e.g. [2,3,4,5,6]). Comparing the target number and a guessing number is allowed. Specifically, determination of these cases are allowed:  
  - 1) target number = guessing number,
  - 2) target number > guessing number,
  - 3) target number < guessing number.

- **Intuition**: using the concept of "divide and conquer," each step truncates an interval into half until the guessing number correctly predicts the target number. For each step, the guessing number is the mid-point of the minimum number and the maximum number of an interval. The min/man numbers are dynamically changed with respect to the relationship of the target number and the guessing number. That is:
  - 1) if target numbmer = guessing number, no need to change min/max.
  - 2) if target number > guessing number, then max = target number.
  - 3) if target number < guessing number, then min = target number.

Except the case of 1), each step is revisited recursively until the target number is discovered.

- **Time complexity**: $O(log_{2}n)$ where $n$ is the number of elements in the given list. $log_{2}$ is related to 'truncating the list into half' each step.

**Example:**
- Target number = 77
- Minimal number = 0
- Maximal number = 100
- How to guess the target number (metric) = mid-point of min and max numbers.

```python
def binarySearch(list):

    # sort the list
    list.sort()

    # create index list
    listIndex = []

    # initialization
    for index in range(len(list)):
      listIndex.append(index)  # example: [0,1,2,3,4,5,6,7,8]

    minIndex = listIndex[0]    # the first index
    maxIndex = listIndex[-1]   # the last index

    import random              # for random.choice()
    targetIndex = random.choice(listIndex)
    guessIndex = int((minIndex+maxIndex)/2) #mid-point of the list

    # display sorted list
    print(f'sorted list: {list}')
    print(f'index of the sorted list: {listIndex}')

    # binary search using the guessing number = mean of an interval
    def binarySearchMid(minIndex, maxIndex, targetIndex, guessIndex, count = 0):

      # keep counting
      count = count + 1

      # comparing targetNumber and guessNumber
      # happy case:
      if targetIndex == guessIndex:
        print(f'count: {count}, target number = guessNumber, target number: {listIndex[targetIndex]}, guess number: {listIndex[guessIndex]}')
        return listIndex[targetIndex]

      elif listIndex[targetIndex] > listIndex[guessIndex]:
        print(f'count: {count}, target number > guessNumber. target number: {listIndex[targetIndex]}, guess number: {listIndex[guessIndex]}')
        minIndex = guessIndex
        guessIndex = int((minIndex + maxIndex)/2)
        return binarySearchMid(minIndex, maxIndex, targetIndex, guessIndex, count)

      elif listIndex[targetIndex] < listIndex[guessIndex]:
        print(f'count: {count}, target number < guessNumber. target number: {listIndex[targetIndex]}, guess number: {listIndex[guessIndex]}')
        maxIndex = guessIndex
        guessIndex = int((minIndex + maxIndex)/2)
        return binarySearchMid(minIndex, maxIndex, targetIndex, guessIndex, count)


    return binarySearchMid(minIndex, maxIndex, targetIndex, guessIndex, count = 0)
```

```python
# testing
import random
randomNumber = random.randint(40, 50)

randomIntList = [random.randint(-100, 100) for index in range(randomNumber)]

result = binarySearch(randomIntList)
print(f'The target number for given list is: {result}')
```
