# Search the largest number in a list

Creation date: July 6, 2024

Example:
- list of numbers: [3,7,1,8,9,2,10,3,7,11,2]
- the largest number of this list is 11

image.png

# Python codes

```python
# search for the largest number in a given list
def searchLargest(list):
    """
    input: list of numbers
    output: a largest number
    intuition:
        1. set the variable 'largest number' as negative infinity.
        2. For each number in the list, if the 'largest number' < this number,
           then update 'largest number' = this number
        3. iterate the step 2 with every element in the given list of numbers
    time complexity: O(n) since each element is visited in the list of numbers
                     where n is the number of elements in the list.
    """
    import math # for math.inf to represent an infinity

    # initialize the largest number
    largestNum = -math.inf

    # iterate for each element in the list
    for index in range(len(list)):
        if list[index] > largestNum:
            largestNum = list[index] # update the larget number

    return largestNum
```

```python
# testing
input = [3,7,1,8,9,2,10,3,7,11,2]
# output should be 11, since it is the largest number in the input
print(f'For the given list of {input}, the largest number in this list is: {searchLargest(input)}') # output: 11
```
